# Checking for outliers in the dataset using the IQR method:

```python
def detect_outliers_iqr(data):
    outlier_summary = {}
    # for column in data.select_dtypes(include=['float64', 'int64']).columns:
    for column in data.select_dtypes(include=['int64']).columns:
        Q1 = data[column].quantile(0.25)
        Q3 = data[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)][column]
        outlier_summary[column] = len(outliers)

        print(f"Column: {column}")
        print(f"  Lower Bound = {lower_bound}")
        print(f"  Upper Bound = {upper_bound}")
        print(f"  Outliers Found = {len(outliers)}\n")

    return outlier_summary

df_forOutlier = df.drop(['Depression', 'Anxiety', 'PanicAttack', 'SpecialistTreatment', 'HasMentalHealthSupport'],axis=1)
outlier_counts = detect_outliers_iqr(df_forOutlier)
```

# Generate target column based on conditions:

```python
df['IsCounselingNeeded'] = (
    (df['Depression'] == 1) |
    (df['Anxiety'] == 1) |
    (df['PanicAttack'] == 1) |
    (df['SpecialistTreatment'] == 1) |
    (df['SymptomFrequency_Last7Days'] >= 3) |
    ((df['HasMentalHealthSupport'] == 0) &
     ((df['SleepQuality'] <= 2) | (df['StudyStressLevel'] >= 3))) |
    (df['CGPA'] <= 2.5) |
    (df['AcademicEngagement'] <= 2)
).astype(int)
```

# Encoding the 'YearOfStudy' as it is the only column where ordinal encoding needed:

```python
X = df.drop('IsCounselingNeeded', axis=1)
y = df['IsCounselingNeeded']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
X_train.shape, X_test.shape
```

```python
categories = [['Year 1', 'Year 2', 'Year 3', 'Year 4']]

oe = OrdinalEncoder(categories=categories)

# Fit only on YearOfStudy
X_train[['YearOfStudy']] = oe.fit_transform(X_train[['YearOfStudy']])
X_test[['YearOfStudy']] = oe.transform(X_test[['YearOfStudy']])
```

```python
# for Course
df['Course'].value_counts()
```

```python
# Count frequency of each course
course_counts = df['Course'].value_counts()

# Replace courses that occur less than 20 times with "Other"
df['Course'] = df['Course'].apply(lambda x: x if course_counts[x] >= 20 else 'Other')

# Now one-hot encode again
# one_hot_encoded_data = pd.get_dummies(df, columns=['Course'], drop_first=True)
one_hot_encoded_data = pd.get_dummies(df, columns=['Course'])

# Convert only the one-hot encoded columns to integer type
for col in one_hot_encoded_data.columns:
    if col.startswith('Course_'):
        one_hot_encoded_data[col] = one_hot_encoded_data[col].astype(int)

print(one_hot_encoded_data.filter(like='Course_').head())
```
