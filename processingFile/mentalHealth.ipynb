Checking for outliers in the dataset using the IQR method.

```python
def detect_outliers_iqr(data):
    outlier_summary = {}
    # for column in data.select_dtypes(include=['float64', 'int64']).columns:
    for column in data.select_dtypes(include=['int64']).columns:
        Q1 = data[column].quantile(0.25)
        Q3 = data[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)][column]
        outlier_summary[column] = len(outliers)

        print(f"Column: {column}")
        print(f"  Lower Bound = {lower_bound}")
        print(f"  Upper Bound = {upper_bound}")
        print(f"  Outliers Found = {len(outliers)}\n")

    return outlier_summary

df_forOutlier = df.drop(['Depression', 'Anxiety', 'PanicAttack', 'SpecialistTreatment', 'HasMentalHealthSupport'],axis=1)
outlier_counts = detect_outliers_iqr(df_forOutlier)
```
